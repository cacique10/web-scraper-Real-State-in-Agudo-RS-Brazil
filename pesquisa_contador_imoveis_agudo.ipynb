```python

lista_tamanhos_m2 = []
lista_valores_m2 = []


# definindo quantas paginas ira procurar, 1,2,3...
for i in range(1,5):
    
    IMOVEL_URL = f"https://www.alineimoveisrs.com.br/imoveis/a-venda/rs/agudo/terreno?&pagina={i}"
    #print("\nURL pagina: ", IMOVEL_URL)
    
    html = requests.get(IMOVEL_URL)
    
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    page = requests.get(IMOVEL_URL, headers=headers)
    soup = BeautifulSoup(page.text)
    
    
    
    ########################################################
    # selecionando elemento com informacao relevante(tamanho)

    tamanho_cont = soup.select_one("div.imovel-icon-item")
    if(tamanho_cont is not None):
        tamanho = tamanho_cont.select_one("span")
        #print(tamanho, tamanho.text)
        for span in soup.select("div.imovel-icon-item span"):
            #print(span, span.text)
            #print(span.text)
            if(span!=0):
                lista_tamanhos_m2.append(span.text)
            else:
                pass


        ########################################################
        # selecionando elemento com informacao relevante(valor)
        valor_cont = soup.select_one("div.definition")
        valor = valor_cont.select_one("span")
        #print(valor, valor.text)
        for div in soup.select("div.definition"):
            #print(div, div.text)
            #print(div.text)
            if('R$' in str(div)):
                lista_valores_m2.append(div.text)
            else:
                pass

        ########################################################
    
cont_teste = 0
# substituindo caracteres indesejados; deuxando apenas numeros
for i in range(0, len(lista_tamanhos_m2)):
    lista_tamanhos_m2[i] =  re.sub('m²','', str(lista_tamanhos_m2[i]))
    lista_tamanhos_m2[i] =  re.sub(' ','', str(lista_tamanhos_m2[i]))
    #print(lista_tamanhos_m2[i])

    
for i in range(0, len(lista_valores_m2)):
    lista_valores_m2[i] =  re.sub('R$','', str(lista_valores_m2[i]))
    lista_valores_m2[i] =  re.sub('\.','', str(lista_valores_m2[i]))
    lista_valores_m2[i] =  lista_valores_m2[i][3:]
    
    #print(lista_valores_m2[i])
    
    
########################################################


# somando todos tamnhos dos terrenos e valores
cont_m2 = 0
cont_valores = 0

for number in lista_tamanhos_m2:
    cont_m2=cont_m2+float(number)

for val in lista_valores_m2:
    cont_valores=float(cont_valores)+float(val)
    

########################################################

print(f"\nForam encontrados {len(lista_tamanhos_m2)} terrenos com metragem em Agudo")

print(f"\n--->Contador m2 somados: {cont_m2:,}")
print(f"Media m2 por terreno: {cont_m2/len(lista_tamanhos_m2):,}")

print(f"\n--->Contador Valores somados: {cont_valores:,}")
print(f"Media de preço por terreno: {cont_valores/len(lista_valores_m2):,}")

print(f"\n--->Terreno mais caro: {max(lista_valores_m2)}")
print(f"--->Maior terreno: {max(lista_tamanhos_m2)} m2")

```
